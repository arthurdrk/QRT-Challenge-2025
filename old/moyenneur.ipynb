```python
import pandas as pd

# Charger les deux fichiers
df1 = pd.read_csv('../submissions/ensemble_with_mtlr.csv')
df2 = pd.read_csv('submissions/7548_moy_7525_7497.csv')

# Vérifier que les IDs sont dans le même ordre
assert all(df1['ID'] == df2['ID'])

# Calculer la moyenne des risk_score
df_mean = df1.copy()
df_mean['risk_score'] = (df1['risk_score'] + 3*df2['risk_score']) / 4

# Sauvegarder le résultat
df_mean.to_csv('submissions/MTLR_moy_ensemble.csv', index=False)
```
