```python
import pandas as pd
from scipy.stats import rankdata

# ===============================
# 1. Charger les prédictions
# ===============================

df1 = pd.read_csv('C:\\Users\\arthr\\Desktop\\ENSAE\\QRT-Challenge-2025\\submissions\\XGB_0.7623.csv')
df2 = pd.read_csv('C:\\Users\\arthr\\Desktop\\ENSAE\\QRT-Challenge-2025\\submissions\\MOY_MTLR_7548.csv')

df3 = pd.read_csv('C:\\Users\\arthr\\Desktop\\ENSAE\\QRT-Challenge-2025\\old\\submissions\\xgb-cindex1-tuned_0.753.csv')
df4 = pd.read_csv('C:\\Users\\arthr\\Desktop\\ENSAE\\QRT-Challenge-2025\\submissions\\XGBCox_0.749.csv')
# ===============================
# 2. Vérifications de sécurité
# ===============================

assert (df1['ID'] == df2['ID']).all(), "IDs df1 / df2 non alignés"
assert (df1['ID'] == df3['ID']).all(), "IDs df1 / df3 non alignés"

# ===============================
# 3. Ranking des risk scores
# ===============================

r1 = rankdata(df1['risk_score'], method='average')
r2 = rankdata(df2['risk_score'], method='average')
r3 = rankdata(df3['risk_score'], method='average')

# ===============================
# 4. Ensembling pondéré
# ===============================

w1, w2, w3 = 0.43, 0.35, 0.30

ensemble_rank = (
    w1 * r1 +
    w2 * r2 +
    w3 * r3
)

# Re-rank final (important)
ensemble_rank = rankdata(ensemble_rank, method='average')

# ===============================
# 5. Sauvegarde soumission finale
# ===============================

submission = pd.DataFrame({
    'id': df1['ID'],
    'risk_score': ensemble_rank
})

submission.to_csv(
    'C:\\Users\\arthr\\Desktop\\ENSAE\\QRT-Challenge-2025\\submissions\\ensemble_weighted_rank_v3.csv',
    index=False
)

print("✅ Soumission ensembliste créée avec succès.")
```

```python

# Calculer la moyenne des risk_score
df_mean = df1.copy()
df_mean['risk_score'] = (df1['risk_score'] + df2['risk_score']) / 2

# Sauvegarder le résultat
df_mean.to_csv('XGBAFT_MTLR_FINAL.csv', index=False)
```
